```python
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        'data/train',
        target_size=(299, 299),
        batch_size=16,
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        'data/validation',
        target_size=(299,299),
        batch_size=16,
        class_mode='binary')
```

```python
x,y = validation_generator.next()
for i in range(0,2):
    image = x[i]
    label = y[i]
    print (label)
    plt.imshow(image)
    plt.show()
```

```python
x,y = train_generator.next()
for i in range(0,2):
    image = x[i]
    label = y[i]
    print (label)
    plt.imshow(image)
    plt.show()
```
